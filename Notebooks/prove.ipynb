```python
import cv2
import numpy as np
from torchvision import transforms
from PIL import Image
import torch
import matplotlib.pyplot as plt
import os
```

```python
im_path = os.path.join("..\\Dataset\\train_set", "train_047253.jpg")

image_original = cv2.imread(im_path)
image_original = cv2.cvtColor(image_original, cv2.COLOR_BGR2RGB)
    
transform = transforms.Compose(
    [
        transforms.ToTensor(),
        transforms.Resize(232, interpolation=transforms.InterpolationMode.BILINEAR),
        transforms.Pad(padding=(0, 0, 0, 0), fill=0),
        transforms.CenterCrop((232, 232)),
    ]
)

image_transformed = transform(image_original)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(image_original)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(image_transformed.permute(1, 2, 0))
axes[1].set_title("Transformed Image")
axes[1].axis("off")

plt.show()
```
